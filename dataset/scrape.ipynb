```python
from seleniumbase import Driver
import time
from selenium.webdriver.common.by import By

USER_DATA_PATH = "/Users/kietho/Library/Application\ Support/Google/Chrome/Default"

driver = Driver(uc=True, user_data_dir=USER_DATA_PATH)
url = 'https://codepen.io/jh3y/pen/KKbpeBQ/'
driver.get(url)

# Click the "Export" button
export_btn = driver.find_element(By.XPATH, "//button[contains(@class, 'Button-module_root-xw+9D') and text()='Export']")
export_btn.click()
time.sleep(1)

# Click the "Export .zip" link
export_zip_link = driver.find_element(By.XPATH, "//a[@data-test-id='export-zip']")
export_zip_link.click()
time.sleep(6)

driver.quit()
```
